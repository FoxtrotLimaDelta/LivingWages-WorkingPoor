```python
total_tanf = {}
fed_tanf = {}
state_tanf = {}
percent_tanf = {}

for i in range(3, sheet.nrows):
    
    row = sheet.row_values(i)
    
    state = row[0]
    basic_assistance = row[1]
    work_activities = row[2]
    work_supports = row[3]
    child_care = row[4]
    adminstration_systems = row[5]
    tax_credit = row[6]
    pre_k = row[7]
    child_welfare = row[8]
    other = row[9]
    total = row[11]
                
    if i in range(3, 55):
            total_tanf[state] = {
                "Basic Assistance": basic_assistance,
                "Work Activities": work_activities,
                "Work Supports": work_supports,
                "Child Care": child_care,
                "Adminstration & Systems": adminstration_systems,
                "Tax Credit": tax_credit,
                "Pre-K": pre_k,
                "Child Welfare": child_welfare,
                "Other": other,
                "Total": total
                    
                }
                
    if i in range(58, 111):                
            fed_tanf[state] = {
                "Basic Assistance": basic_assistance,
                "Work Activities": work_activities,
                "Work Supports": work_supports,
                "Child Care": child_care,
                "Adminstration & Systems": adminstration_systems,
                "Tax Credit": tax_credit,
                "Pre-K": pre_k,
                "Child Welfare": child_welfare,
                "Other": other,
                "Total": total
                }
         
    if i in range(113, 166):
                state_tanf[state] = {
                "Basic Assistance": basic_assistance,
                "Work Activities": work_activities,
                "Work Supports": work_supports,
                "Child Care": child_care,
                "Adminstration & Systems": adminstration_systems,
                "Tax Credit": tax_credit,
                "Pre-K": pre_k,
                "Child Welfare": child_welfare,
                "Other": other,
                "Total": total
                }
    
    if i in range(168, 220):
                percent_tanf[state] = {
                "Basic Assistance": 'basic_assistance',
                "Work Activities": 'work_activities',
                "Work Supports": 'work_supports',
                "Child Care": 'child_care',
                "Adminstration & Systems": 'adminstration_systems',
                "Tax Credit": 'tax_credit',
                "Pre-K": 'pre_k',
                "Child Welfare": 'child_welfare',
                "Other": 'other',
                "Total": 'total'
                }
            
    print (i,row)
```
