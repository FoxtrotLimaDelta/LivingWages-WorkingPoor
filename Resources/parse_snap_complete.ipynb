```python
import xlrd
import pandas as pd
import pprint
```

## SNAP 2019

```python
book = xlrd.open_workbook('SNAP/FY19.xls')
```

```python
for sheet in book.sheets():
    print (sheet.name)
```

### NERO

```python
sheet = book.sheet_by_name('NERO')
```

```python
for i in range(sheet.nrows):
    print(i,sheet.row_values(i))
```

```python
snap_data_2019 = {}
for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [x for x in range(7, sheet.nrows, 15)]:
        #for the rows that have the state names
        state = row[0]
        if state == 'NERO':
            break
        else:
            continue         
    if i in [x for x in range(20, sheet.nrows, 15)]:
        
        snap_data_2019[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }
```
