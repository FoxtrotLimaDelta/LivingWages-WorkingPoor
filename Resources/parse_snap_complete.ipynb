## SNAP 2019

```python
book = xlrd.open_workbook('SNAP/FY19.xls')
```

```python
for sheet in book.sheets():
    print (sheet.name)
```

### NERO

```python
sheet = book.sheet_by_name('NERO')
```

```python
for i in range(sheet.nrows):
    print(i,sheet.row_values(i))
```

```python
snap_data_2019 = {}
for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [x for x in range(7, sheet.nrows, 15)]:
        #for the rows that have the state names
        state = row[0]
        if state == 'NERO':
            break
        else:
            continue         
    if i in [x for x in range(20, sheet.nrows, 15)]:
        
        snap_data_2019[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }
```

```python
pprint.pprint(snap_data_2019)
```

```python
snap_data = {}

########################
#                      #
# NERO                 #
#                      #
########################

sheet = book.sheet_by_name('NERO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [x for x in range(7, sheet.nrows, 15)]:
        #for the rows that have the state names
        state = row[0]
        if state == 'NERO':
            break
        else:
            continue         
    if i in [x for x in range(20, sheet.nrows, 15)]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }

########################
#                      #
# MARO                 #
#                      #
########################

sheet = book.sheet_by_name('MARO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [7,22,37,52,67,82,84,99,114,129]:
        #for the rows that have the state names
        state = row[0]
        if state == 'MARO':
            break
        else:
            continue       
    if i in [20,35,50,65,80,97,112]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }

########################
#                      #
# SERO                 #
#                      #
########################

sheet = book.sheet_by_name('SERO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [7,22,37,52,67,82,97,112,127,142]:
        #for the rows that have the state names
        state = row[0]
        if state == 'SERO':
            break
        else:
            continue       
    if i in [20, 35, 50, 65, 80, 95, 110, 125]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }

########################
#                      #
# MWRO                 #
#                      #
########################

sheet = book.sheet_by_name('MWRO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [7,22,37,52,67,82,97,112,127,142]:
        #for the rows that have the state names
        state = row[0]
        if state == 'MWRO':
            break
        else:
            continue       
    if i in [20, 35, 50, 65, 80, 95, 110, 125]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }

########################
#                      #
# SWRO                 #
#                      #
########################

sheet = book.sheet_by_name('SWRO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [7,22,37,52,67,82,97,112,127,142]:
        #for the rows that have the state names
        state = row[0]
        if state == 'SWRO':
            break
        else:
            continue       
    if i in [20, 35, 50, 65, 80, 95, 110, 125]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }

########################
#                      #
# MPRO                 #
#                      #
########################
sheet = book.sheet_by_name('MPRO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [7,22,37,52,67,82,97,112,127,142]:
        #for the rows that have the state names
        state = row[0]
        if state == 'MPRO':
            break
        else:
            continue       
    if i in [20, 35, 50, 65, 80, 95, 110, 125]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }


########################
#                      #
# WRO                  #
#                      #
########################
sheet = book.sheet_by_name('WRO')

for i in range(7, sheet.nrows):#Start at 7 b/c data starts here
    
    #RULES: 
    
    row = sheet.row_values(i)
    
    if i in [7,22,37,52,67,82,97,112,127,142]:
        #for the rows that have the state names
        state = row[0]
        if state == 'WRO':
            break
        else:
            continue       
    if i in [20, 35, 50, 65, 80, 95, 110, 125]:
        
        snap_data[state] = {
        'household_participation': row[1],
        'individual_participation': row[2],
        'total_cost' : row[3],
        'cost_by_household' : row[4],
        'cost_by_individual': row[5],
        }

```
