## 2014 TANF

```python
book = xlrd.open_workbook('TANF/tanf_financial_data_fy_2014.xlsx')
```

```python
for sheet in book.sheets():
    print (sheet.name)
```

```python
sheet = book.sheet_by_name('Federal Assistance')
```

```python
for i in range(sheet.nrows):
    print(i,sheet.row_values(i))
```

### Federal Assistance

```python
fed_tanf_2014 = {}
for i in range(4, sheet.nrows):
    
    row = sheet.row_values(i)
    
    state = row[0]
    
    fed_tanf_2014[state] = {
        'Basic Assistance' : row[2],
        'Child Care Assistance' : row[3],
        'Transportation and Supportive Services': row[4],
        'Assistance Under Prior Law': row[5]
    }
    
```

### Federal Non-Assistance

```python
sheet = book.sheet_by_name('Federal Non-Assistance')
```

```python
for i in range(sheet.nrows):
    print(i,sheet.row_values(i))
```

```python
fed_tanf_2014_2 = {}
for i in range(4, sheet.nrows):
    row = sheet.row_values(i)
    
    state = row[0]
    
    fed_tanf_2014_2[state] = {
        'Work Related Activities': row[2],
        'Child Care Non-Assistance': row[3],
        'Transportation': row[4],
        'Individual Development Accounts': row[5],
        'Refundable Earned Income Tax Credits': row[6],
        'Non-EITC Refundable State Tax Credits': row[7],
        'Non-Recurrent Short Term Benefits': row[8],
        'Prevention of Out-of-Wedlock Pregnacies': row[9],
        'Family Formation and Maintence Programs': row[10],
        'Administrative Costs': row[11],
        'Systems': row[12],
        'Non-Assistance Under Prior Law': row[13],
        'Other': row[14]
    }
```

```python
for key in fed_tanf_2014:
    if key in fed_tanf_2014_2:
        fed_tanf_2014[key].update(fed_tanf_2014_2[key])
```

```python
fed_tanf_2014_df = pd.DataFrame.from_dict(fed_tanf_2014, orient = "index")
```

```python
pd.DataFrame.to_csv(fed_tanf_2014_df,'TANF/TANF DFs/fed_2014_df.csv')
```
